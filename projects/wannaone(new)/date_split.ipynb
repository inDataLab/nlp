## 파일을 읽습니다

```python
# 특정 디렉토리 아래의 모든 파일의 절대경로 목록을 리턴하는 함수
import os
import pandas as pd

def allfiles(path):
    res = []
    for root, dirs, files in os.walk(path):
        rootpath = os.path.join(os.path.abspath(path), root)
        for file in files:
            filepath = os.path.join(rootpath, file)
            if 'xlsx' in filepath:
                res.append(filepath)
    return res

curDir = os.getcwd()

######################################
############# INPUT HERE #############
######################################

# 있는 파일 다 불러올꺼
path = '/Users/EternalFlow/Desktop/indatalab/nlp/projects/wannaone(new)/result_data/twit_170811_wannaone'
files = allfiles(path)
```

# 파일을 씁니다

```python
for i in files:
    path = i
    original_file = path[path.rfind('/')+1:path.rfind('.')]
    df=pd.read_excel(path,encoding='utf-8')
    date = df['date'].unique()
    writer = pd.ExcelWriter(curDir+'/date_split/'+original_file+'.xlsx')
    for j in date:
        subset = pd.DataFrame(data=df[df['date']==j])
        subset.to_excel(writer,j)
        print(original_file, j)
    writer.save()
```
