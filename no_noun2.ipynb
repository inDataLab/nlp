```python
import pandas as pd
import numpy as np
# train data를 df로
df_train = pd.read_excel(r'.\projects\wannaone\remixed.xlsx')
# data.frame을 NumPy 배열로
# train set
text_train = df_train['contents'].as_matrix()
y_train = df_train['label'].as_matrix()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression as lgt
from hanspell import spell_checker

# tokeinize - Twitter
from konlpy.tag import Twitter
twit = Twitter()
words = ['Verb','Adjective','Adverb','Conjunction','Exclamation','KoreanParticle','Eomi','Hashtag']
stopwords = ['./Punctuation', ',/Punctuation']

def tokenize(doc):
    tagged = ['/'.join(t) for t in twit.pos(doc, norm=True, stem=True)]
    result = [word for word in tagged if word.split('/')[-1] in words]
    return result

tfidf = TfidfVectorizer(tokenizer=tokenize,ngram_range=(1,3),min_df=3,stop_words=stopwords)
```

```python
# tokeinize - Kkma
from konlpy.tag import Kkma
kkma = Kkma()
words = ['NNG','VXV','VXA','VCP','VCN','MAG','MAC','IC','SF','SE','EMO']
stopwords = ['./Punctuation', ',/Punctuation']
def tokenize(doc):
    tagged = ['/'.join(t) for t in kkma.pos(doc)]
    result = [word for word in tagged if word.split('/')[-1] in words]
    return result
tfidf2 = TfidfVectorizer(tokenizer=tokenize,ngram_range=(1,3),min_df=3,stop_words=stopwords)
```

```python
input_data = tfidf.fit_transform(text_train.astype('U'))
clf_1 = lgt(C=10)
print('ML Start')
model = clf_1.fit(input_data,y_train)
print('ML Finish')
```
